# **MRI Musculoskeltal Parameter Reference Values:**

This document reflects to some extent the geometric, contrast and NSA values on predominantly 1.5 Tesla Philips scanners $(\small 30 \text{ mT/m}; 150 \text{ sl rate})$. The values change as changes are introduce to improve the ratio of signal-resolution to scanning time, and in this regard, it is a continued work in progress. Further, these parameters are orientative since anatomical and coverage considerations my require changes to address conflict warnings: e.g. consider increasing the TR values to control time when anatomical coverage increases the number of packages.

---

Notice that when working on Philips magnets the pixel size is preserved by default, and hence, the matrix changes as FoV varies. Therefore, the in-plane geometry is best controlled directly establishing the voxel size, explaining the pursposeful omission of acquisition and reconstruction matrix values below.

---

In Philips magnets the receiver bandwidth is given in WFS (pixels/KHz). The conversion to KHz used by other vendors is

$$\small \text{Bandwidth [KHz]}=K \;\frac{\text{Matrix (Freq)}}{\text{WFS}}$$

with $K$ being the WFS conversion factor, which is $0.11$ for $1.5$-Tesla units (for example).
